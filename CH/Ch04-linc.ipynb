- Comparsion

|Comparison|Operator|True example|False Example|
--|--|--|--
|Less than |<| 2 < 3 |2 < 2| 
| Greater than | > | 3>2 | 3>3 | 
| Less than or equal | <= | 2 <= 2 | 3 <= 2 |  
| Greater or equal | >= | 3 >= 3 | 2 >= 3 | 
| Equal | == | 3 == 3 | 3 == 2 | 
| Not equal | != | 3 != 2 | 2 != 2 | 

```python
1 < 1 + 1 < 3
```

```python
x = 12
y = 5
min(x, y) <= (x+y)/2 <= max(x, y)
```

```python
"Dog" > "Catastrophe" > "Cat"
```
