### 문제1
2명의 사람이 가위바위보를 100번 시행하여 비긴 횟수를 리턴하는 함수 count_tie_100()를 작성하시오. 

(단, 위에서 정의한 rcp 배열을  이용하여 작성할 것)

```
### 답안을 작성하시오.
def count_tie_100():
  a = np.random.choice(rcp, 100)
  b = np.random.choice(rcp, 100)
  x = np.count_nonzero( a==b )
  return (x)
```

```
### 지우지 마시오.
count_tie_100()
```

### 문제2
2명의 사람이 가위바위보를 ``no_rcp``번 시행하여 비긴 횟수를 리턴하는 함수 count_tie()를 작성하시오. 

(단, ``no_rcp``은 입력 파라미터이며 default는 100으로 설정한다.)

```
### 답안을 작성하시오.
def count_tie(no_rcp=100):
  a = np.random.choice(rcp, no_rcp)
  b = np.random.choice(rcp, no_rcp)
  x = np.count_nonzero( a==b )
  return (x)
```

```
count_tie(500)
```

```
count_tie()
```

### 문제3
2명이 ``no_rcp``번 랜덤하게 가위바위보 했을 때 비긴 ***비율***(주의: 횟수가 아님)을 조사하는 시행을 ``no_trial``번 반복하여 그 결과(비긴 비율)를 배열로 리턴하는 함수 ratio_tie를 작성하시오.

(``no_rcp``과 ``no_trial``은 입력 파라미터이며 default값은 100, 10000으로 각각 설정한다.)

```
### 답안을 작성하시오.
def ratio_tie(no_rcp=100, no_trial=10000):
  result = make_array()
  for i in np.arange(no_trial):
    x = count_tie(no_rcp)
    result = np.append(result, x)  
  return (result/no_rcp)
```

```
ratio_tie(50, 5000)
```

```
ratio_tie()
```

### 문제4
위 ratio_tie함수를 호출후 그 결과(배열)를 리턴받아  아래 출력 결과와 같도록 Table 변수 ``simulation_results``에 저장하시오.

(단, ``no_rcp``과 ``no_trial``는 default값으로 호출)

```
### 답안을 작성하시오.
num_repetitions = 10000
simulation_results = Table().with_column(
    'Repetition', np.arange(1, num_repetitions + 1),
    'Ratio of Ties', ratio_tie()
)
```
