```
np.random. seed(0)
def rand_table(r, n=1000, mx=0, my=0, s=1):
    x = mx + s * np.random.normal(0, 1, n)
    z = np.random.normal(0, 1, n)
    y = r*x*s + (np.sqrt(1-r**2))*z
    #plots.scatter(x, y)
    t= Table().with_columns('Y', y, 'X', x)
    return t
dt = rand_table(0.5, 1000, 2, 4, 2)   
dt
```
